# HeatMaps

## Importing required libraries

```python
import pandas as pd

import plotly.graph_objs as go
import plotly.offline as offline

offline.init_notebook_mode(connected=True)
```

## Heatmap plot

```python
trace = go.Heatmap(z = [[8,64],[3,37]])

data = [trace]

offline.iplot(data)
```

## Load some data

```python
df = pd.read_xml('Datasets/cal_housing.xml')
```

## Check dimensions

## Select a random samples from data

```python
df = df.sample(frac=0.1).reset_index(drop=True)
```

```python
df.head()
```

```python
df.shape
```

## Categorize house values by location

```python
trace =  go.Heatmap(z = df['median_house_value'], y = df['longitude'], x = df['latitude'], reversescale=True)

data = [trace]

offline.iplot(data)
```
