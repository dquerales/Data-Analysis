# Categorical Plots

## Importing required libraries

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

## Loading the Data

```python
df = pd.read_csv('Datasets/tips.csv')
```

```python
df.head()
```

```python
df.shape
```

## Bar Plot

A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally. A vertical bar chart is sometimes called a column chart.

```python
plt.figure(figsize=(10,8))

sns.barplot(x = 'size', y = 'total_bill', data = df)

plt.show()
```

Black bars are errors  estimates 

## Count plot

Show the counts of observations in each categorical bin using bars.

```python
plt.figure(figsize=(10,8))

sns.countplot(x = 'size', data = df)

plt.show()
```

## Group Bar graph

```python
plt.figure(figsize=(10,8))

sns.barplot(x = 'size', y = 'total_bill', hue ='smoker', data = df)

plt.show()
```

### Add Standard deviation

In statistics, the standard deviation is a measure of the amount of variation or dispersion of a set of values. A low standard deviation indicates that the values tend to be close to the mean of the set, while a high standard deviation indicates that the values are spread out over a wider range.

![sd](https://assets.ltkcontent.com/images/83660/calculating-standard-deviation_0066f46bde.jpg)

```python
plt.figure(figsize=(10,8))

sns.barplot(x = 'size', y = 'total_bill', data = df, ci='sd')

plt.show()
```

### Look for couples

```python
df['couple'] = df['size'] == 2
```

```python
df.sample(10)
```

```python
plt.figure(figsize=(10,8))

sns.barplot(x = 'size', y = 'total_bill', data = df, hue='couple', dodge='False')

plt.show()
```

## Point Plot

A point plot represents an estimate of central tendency for a numeric variable by the position of scatter plot points and provides some indication of the uncertainty around that estimate using error bars.

```python
plt.figure(figsize=(10,8))

sns.pointplot(x = 'size', y = 'total_bill', data = df)

plt.show()
```

### Add hue

```python
plt.figure(figsize=(10,8))

sns.pointplot(x = 'size', y = 'total_bill', hue = 'sex', data = df)

plt.show()
```

## Cat plot

### Use strip plot as default

```python
sns.catplot(x = 'size', y = 'total_bill', hue = 'time', data = df, height=5, aspect=1.5)

plt.show()
```

### As Box plot

```python
sns.catplot(x = 'day', y = 'total_bill', kind = 'box', data = df, aspect=1.5)

plt.show()
```

### As violin

```python
sns.catplot(x = 'size', y = 'total_bill', hue ='time', kind = 'violin', data = df, height=5, aspect=1.5)

plt.show()
```
