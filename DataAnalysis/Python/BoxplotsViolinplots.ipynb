# BoxPlots & ViolinPlots

## Importing required libraries

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

## Loading the Data

```python
df = pd.read_csv('Datasets/Admission_Predict.csv')

df.head()
```

```python
df.shape
```

## Box Plot

In descriptive statistics, a box plot or boxplot (also known as box and whisker plot) is a type of chart often used in explanatory data analysis. Box plots visually show the distribution of numerical data and skewness through displaying the data quartiles (or percentiles) and averages.

![Box plot](https://www.simplypsychology.org/boxplot.jpg)

### One variable

```python
plt.figure(figsize=(10,8))

sns.boxplot(x = 'GRE Score', data = df)

plt.show()
```

#### Vertical

```python
plt.figure(figsize=(10,8))

sns.boxplot(y = 'GRE Score', data = df)

plt.show()
```

### Two variables

```python
plt.figure(figsize=(10,8))

# Gre score by University Rating
sns.boxplot(x = 'University Rating', y = 'GRE Score', data = df)

plt.show()
```

### Three variables

```python
plt.figure(figsize=(10,8))

# Gre score by University Rating
sns.boxplot(x = 'University Rating', y = 'GRE Score', hue = 'Research', data = df, palette='Set1')

plt.show()
```

## Violin Plots

A violin plot is a method of plotting numeric data. It is similar to a box plot, with the addition of a rotated kernel density plot on each side.

### One variable

```python
plt.figure(figsize=(10,8))

sns.violinplot(x = 'GRE Score', data = df)

plt.show()
```

### Two variables

```python
plt.figure(figsize=(10,8))

# Gre score by University Rating
sns.violinplot(x = 'University Rating', y = 'GRE Score', data = df)

plt.show()
```

### Three variables

```python
plt.figure(figsize=(10,8))

# Gre score by University Rating
sns.violinplot(x = 'University Rating', y = 'GRE Score', hue = 'Research', data = df, palette='Set2')

plt.show()
```

### Merge category

```python
df['First Class'] = df['CGPA'] > 8.5
```

```python
df.head()
```

```python
plt.figure(figsize=(10,8))

# Gre score by University Rating
sns.violinplot(x = 'University Rating', y = 'GRE Score', hue = 'First Class', data = df, dodge=False)

plt.show()
```

## Multiples types of plots

```python
plt.figure(figsize=(10,8))

# Gre score by University Rating
sns.boxplot(x = 'University Rating', y = 'GRE Score', data = df, whis=np.inf)
sns.swarmplot(x = 'University Rating', y = 'GRE Score', data = df, color='black')

plt.show()
```
