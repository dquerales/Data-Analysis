# BubbleMaps

## Importing required libraries

```python
import pandas as pd

import plotly.graph_objs as go
import plotly.offline as offline

offline.init_notebook_mode(connected=True)
```

## Scattergeo plot

```python
trace = dict(type='scattergeo', lon = [-97.92, 0, 78.8], lat = [39.3, 0, 21.76],
            marker = dict(size=10), mode='markers')

data = [trace]

layout = dict(showlegend=False, geo=dict(showland=True))

fig = dict(data = data, layout = layout)

offline.iplot(fig)
```

## Select a random samples from data

```python
df = df.sample(frac=0.1).reset_index(drop=True)
```

```python
df.shape
```

## Geolocation Map

```python
trace = dict(type='scattergeo', lon = df['longitude'], lat = df['latitude'],
            marker = dict(size=df['median_house_value']/1000, sizemode='area'), mode='markers')

data = [trace]

layout = dict(showlegend=False, geo=dict(showland=True))

fig = dict(data = data, layout = layout)

offline.iplot(fig)
```
