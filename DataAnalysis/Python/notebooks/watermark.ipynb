```python
import numpy as np
import matplotlib.cbook as cbook
import matplotlib.image as image
import matplotlib.pyplot as plt
```

```python
with cbook.get_sample_data('/content/logo.png') as file:
    img = image.imread(file)

fig, ax = plt.subplots()
ax.plot(np.sin(np.linspace(0, 10)))
ax.grid()
fig.figimage(img, 420, 40, alpha=0.1)
plt.show()
```
