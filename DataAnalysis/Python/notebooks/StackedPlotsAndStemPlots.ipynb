# StackedPlotsAndStemPlots

## Importing required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Loading the Data

```python
df = pd.read_csv('Datasets/annual_visitation_by_park.csv', skiprows=3, low_memory=False)
```

source: https://irma.nps.gov/STATS/Reports/Park

## Data Transformation 

```python
df.columns
```

### Keep only Park Name, Year & Visits

```python
df = df[['Rowelement1', 'Colhead2', 'Rowelement2']]
```

```python
df.head()
```

### Rename columns

```python
new_names =  {'Rowelement1': 'Name',
              'Colhead2': 'Year',
              'Rowelement2': 'Visits'}
df.rename(columns=new_names, inplace=True)
df.head()
```

```python
df['Name'].unique()
```

### Keep only Badlands, Grand Canyon & Bryce Canyon 

```python
df = df[(df['Name'] == 'Badlands NP')|(df['Name'] == 'Grand Canyon NP')|(df['Name'] == 'Bryce Canyon NP')]
```

```python
df.sample(10)
```

### Drop nulls

```python
df.shape
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df.head(30)
```

### Check data types

```python
df.dtypes
```

#### Change Year, Visits to int

```python
df['Year'] = df['Year'].astype('int')
```

## Pivot table

```python
df = df.pivot(index= 'Year', columns= 'Name', values='Visits')
df
```

```python
df = df.replace(',','', regex=True)
df
```

```python
df['Badlands NP'] = pd.to_numeric(df['Badlands NP'])
df['Grand Canyon NP'] = pd.to_numeric(df['Grand Canyon NP'])
df['Bryce Canyon NP'] = pd.to_numeric(df['Bryce Canyon NP'])
```

```python
df.dtypes
```

```python
df
```

## Stacked Plot

### Extract for Numpy

```python
X = df.index
y = np.vstack([df['Badlands NP'],df['Grand Canyon NP'],df['Bryce Canyon NP']])
```

```python
labels = ['Badlands','Grand Canyon','Bryce Canyon']
colors = ['sandybrown', 'tomato', 'skyblue']
```

```python
plt.figure(figsize=(12,8))
plt.stackplot(X,y, labels=labels, colors=colors)

plt.legend()
plt.show()
```

## Stem plot

### Difference of visitor compared previous year

```python
df[['Badlands NP','Grand Canyon NP','Bryce Canyon NP']] = df[['Badlands NP','Grand Canyon NP','Bryce Canyon NP']].diff()
df.head()
```

```python
plt.figure(figsize=(12,8))

plt.stem(X,df['Grand Canyon NP'], use_line_collection=True)

plt.title('Change in Number of visitors')
plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.stem(X,df['Bryce Canyon NP'], markerfmt = 'g_', linefmt = 'r--', basefmt = 'b:', use_line_collection=True)

plt.title('Change in Number of visitors')
plt.show()
```
