# Data Transformation, Clean & Explore

## Import

```python
import pandas as pd
import numpy as  np
```

## Extract the data

```python
cars = pd.read_csv('Datasets/cars.csv', header=None, delim_whitespace=True)
```

```python
cars.head()
```

## Rename Columns 

```python
new_names =  {0: 'mpg',
              1: 'cylinders',
              2: 'displacement',
              3: 'horsepower',
              4: 'weight',
              5: 'acceleration',
              6: 'model_year',
              7: 'origin',
              8: 'car_name'}
```

```python
cars.rename(columns=new_names, inplace=True)
cars.head()
```

## Check dimensions 

```python
cars.shape
```

### Check at Random 

```python
cars.sample(5)
```

```python
cars.tail(5)
```

## Explore the data

### Acceleration vs Displacement (also HP)

```python
cars.plot.scatter(x = 'displacement', y = 'acceleration',
                  c = 'horsepower',
                  colormap = 'viridis',
                  figsize = (10,8));
```

greater displcement, lower acceleration, more hp

### MPG vs Weight

```python
cars.plot.scatter(x = 'mpg', y = 'weight',
                  figsize=(10,8));
```

greater the mpg lower the weight

### Model year vs Mpg

```python
cars.plot.scatter(x='model_year', y='mpg', figsize=(10,8))
```

greater the year more mpg

## Check Nulls

```python
cars.isnull().sum()
```

there is 8 values of mpg nulls

### Replace the nulls with mean

```python
cars['mpg']  = cars['mpg'].fillna(cars['mpg'].mean())
```

```python
cars.isnull().sum()
```

```python
cars['model_year'].dtype
```

## Horse Power vs MPG

```python
cars.plot.scatter(x='horsepower', y='mpg', figsize=(10,8))
```

greater hp lower mpg

```python
cars.plot.scatter(x = 'horsepower', y = 'mpg',
                  c = 'cylinders',
                  colormap = 'viridis',
                  figsize = (10,8));
```

greater hp, lower mpg, higher cylinders
