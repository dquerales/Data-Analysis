# Categorical Data: StripPlots & SwampPlots

## Importing required libraries

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

## Loading the Data

```python
df = pd.read_csv('Datasets/tips.csv')

df.head()
```

```python
df.shape
```

## Strip Plot

Draw a scatterplot where one variable is categorical.

A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

```python
sns.stripplot(x = 'sex', y = 'tip', data = df)

plt.show()
```

male gives more tips

### Disable jitter

```python
sns.stripplot(x = 'sex', y = 'tip', data = df, jitter=False)

plt.show()
```

### Day vs Tip

```python
sns.stripplot(x = 'day', y = 'tip', data = df, jitter=0.04)

plt.show()
```

saturday is a good day for tips

## Swarm Plots

A swarm plot is another way of plotting the distribution of an attribute or the joint distribution of a couple of attributes.

### One variable

```python
plt.figure(figsize=(10,8))

sns.swarmplot(x = 'tip', data = df)

plt.show()
```

### Two variables

```python
plt.figure(figsize=(10,8))

sns.swarmplot(x = 'size', y = 'tip', data = df)

plt.show()
```

### Another dimension

```python
plt.figure(figsize=(10,8))

sns.swarmplot(x = 'size', y = 'tip', hue = 'sex', data = df)

plt.show()
```

men are usually outliers

#### Better visualization

```python
plt.figure(figsize=(10,8))

sns.swarmplot(x = 'size', y = 'tip', hue = 'sex', data = df, palette='Set1', dodge=True)

plt.show()
```
