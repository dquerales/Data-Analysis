# Regression Plots: Lmplot & Replot

## Importing required libraries

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

## Loading the Data

```python
df = pd.read_csv('Datasets/tips.csv')

df.head()
```

```python
df.shape
```

## LM Plot

Draw a scatter data and a linear regression model fit.

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, aspect=1.5)

plt.show()
```

### Add a hue

Grouping variable that will produce points with different colors. Can be either categorical or numeric, although color mapping will behave differently in latter case. Assigning a variable to hue will map its levels to the color of the points.

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue ='smoker', data = df, aspect=1.5)

plt.show()
```

No smokers pays more tips

### Size vs Tip

```python
sns.lmplot(x = 'size', y = 'tip', data = df, aspect=1.5)

plt.show()
```

### Add Jitter

Add uniform random noise of this size to either the x or y variables. The noise is added to a copy of the data after fitting the regression, and only influences the look of the scatterplot. This can be helpful when plotting variables that take discrete values.

```python
sns.lmplot(x = 'size', y = 'tip', data = df, aspect=1.5, x_jitter=0.2)

plt.show()
```

### Add col

Variables that define subsets of the data, which will be drawn on separate facets in the grid. 

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, col='smoker')

plt.show()
```

### Add hue & col

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df, col='day', hue='day', height = 5, aspect= 0.8)

plt.show()
```

## Reg Plots

Plot data and a linear regression model fit. has more flexibility for user inputs, can accept arrays and pandas series objects

```python
plt.figure(figsize=(10,8))

sns.regplot(x = 'total_bill', y = 'tip', data = df)

plt.show()
```

```python
plt.figure(figsize=(10,8))

sns.regplot(df['total_bill'], df['tip'], color = 'b', marker='*')

plt.show()
```

## Plot for logistic models

### Add a column to Big Tips

```python
df['Big tip'] = (df.tip / df.total_bill) > 0.175
```

```python
df.sample(10)
```

```python
plt.figure(figsize=(10,8))

sns.regplot(x = 'total_bill', y = 'Big tip', data = df, logistic=True, y_jitter=0.01)

plt.show()
```
