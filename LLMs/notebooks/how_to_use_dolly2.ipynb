## Import dependencies

```python
import torch
from transformers import pipeline
```

```python
generate_text = pipeline(model="databricks/dolly-v2-2-8b", torch_dtype=torch.bfloat16, trust_remote_code=True, device_map="auto")
```

## Ask


```python
%%time
generate_text("how much protein i need for 70Kg of weight and 1.7 meters of height")
```
